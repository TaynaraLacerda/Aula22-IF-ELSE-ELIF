```python
# Atividade 1 — Classificação de número
## Objetivo: usar if e else. Crie um programa que:Peça um número. Verifique se ele é:
## Positivo / Negativo

numero = float(input("Digite um núemro:"  ))
if numero >0:
    print("Número positivo")
elif numero == 0:
    print("Número Nulo")
else:
    print("Número negativo")
```

```python
# Atividade 2 — Par ou Ímpar. Objetivo: introduzir lógica com operador.
## Crie um programa que: Peça um número. Verifique se ele é: Par / Ímpar

numero = int(input("Digite um número:"))
if  numero  % 2 == 0:
    print("O número é par")
else:
    print("O número é impar")
```

```python
# Atividade 3 — Sistema de notas. Objetivo: usar elif
## Crie um programa que: Peça a nota de um aluno. Classifique:

notas = float(input("Digite uma nota:  "))
if notas >= 7:
  print("Aprovado")
elif notas >= 5:
  print("Recuperação")
else:
    print("Reprovado")
```

```python
# Atividade 4 — Simulação de login simples
## Objetivo: aplicar condição mais real. Crie um programa que: Peça: usuário / senha. Verifique:

usuario = input("Digite o seu loguin:  ")
senha = int(input("Digite a sua senha:  "))
if usuario == "admin" and senha == 1234:
  print("Login realizado com sucesso")
else:
    print("Usuário ou senha inválidos")
```
